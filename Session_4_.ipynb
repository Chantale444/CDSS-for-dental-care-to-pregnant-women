```python
#A function that promts for 2 numbers and calculates
# the sum of the number if they are odd
def sum_odd_between(a, b):
    total = 0
    for i in range(a, b + 1):  # +1 för att inkludera b
        if i % 2 == 1:
            total += i
    return total


# Prompt user
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

# Se till att intervallet funkar även om användaren skriver större först
start = min(num1, num2)
end = max(num1, num2)

result = sum_odd_between(start, end)
print("Sum of odd numbers:", result)
```

```python
# A program that takes as input a list and outputs
#average score, median, mode,
def student_grades():
    n = int(input("How many students? "))

    students = []

    # Input names and scores
    for i in range(n):
        name = input("Enter student name: ")
        score = float(input("Enter score: "))

        students.append({"name": name, "score": score})

    # Initialize variables
    total = 0
    highest_score = students[0]["score"]
    top_student = students[0]["name"]
    scores = []

    # Loop through students
    for student in students:
        score = student["score"]
        total = total + score
        scores.append(score)

        # Check highest score
        if score > highest_score:
            highest_score = score
            top_student = student["name"]

    # Average (mean)
    average = total / n

    # Median
    scores.sort()
    if n % 2 == 0:
        median = (scores[n//2 - 1] + scores[n//2]) / 2
    else:
        median = scores[n//2]

    # Mode (simple version)
    mode = scores[0]
    max_count = 0

    for s in scores:
        count = scores.count(s)
        if count > max_count:
            max_count = count
            mode = s

    # Students above 80
    print("\nStudents scoring above 80:")
    for student in students:
        if student["score"] > 80:
            print(student["name"])

    # Final results
    print("\nAverage:", average)
    print("Highest score:", highest_score)
    print("Top student:", top_student)
    print("Median:", median)
    print("Mode:", mode)


student_grades()
```
